```python
import pandas as pd
import numpy as np

special_rank = pd.read_json('special_rank.json')
universe_special = pd.read_json('universe_special1.json')
# special_ = pd.read_csv('special_id - 副本.txt')
# for i in range(0, len(special_)):
#     print(special_[i])
```

```python
universe_special.loc[:, 'school_id'] = universe_special.loc[:, 'school_id'].astype('int64').astype('str')
special_rank.loc[:, 'school_id'] = special_rank.loc[:, 'school_id'].astype('str')
```

```python
universe_special.rename(columns = {'special_name':'special_name_copy'}, inplace = True)
```

```python
universe_special.loc[:, 'special_name'] = universe_special.loc[:, 'special_name_copy'].apply(lambda x: x.split('（')[0] if '（' in x else x)
df = pd.merge(universe_special, special_rank, how='left', on=['special_name', 'school_id'])
df.drop(columns='special_name', inplace=True)
df.rename(columns={'special_name_copy': 'special_name'}, inplace=True)
df
```

```python
df.columns = ['ID', 'fk_university_id', 'name', 'degree', 'mainBranch', 'subBranch', 'duration', 'detail', 'careerInfo', 'id', 'ruanKeScore' ]
df
```

```python
# ['ID', 'fk_university_id', 'name', 'degree', 'mainBranch', 'subBranch', 'duration', 'detail', 'careerInfo', 'id', 'ruanKeScore' ]
major = pd.DataFrame({
    'id': df.loc[:, 'ID'],
    'name': df.loc[:, 'name'],
    'mainBranch': df.loc[:, 'mainBranch'],
    'subBranch': df.loc[:, 'subBranch'],
    'duration': df.loc[:, 'duration'],
    'careerInfo': df.loc[:, 'careerInfo']
})
major = major.drop_duplicates(subset=['id'], keep='first')
major.to_json('major.json')
```

```python
# ['ID', 'fk_university_id', 'name', 'degree', 'mainBranch', 'subBranch', 'duration', 'detail', 'careerInfo', 'id', 'ruanKeScore' ]
rel_major_university = pd.DataFrame({
    'fk_university_id': df.loc[:, 'fk_university_id'],
    'fk_major_id': df.loc[:, 'ID'],
    'degree': df.loc[:, 'degree'],
    'ruanKeScore': df.loc[:, 'ruanKeScore'],
    'detail': df.loc[:, 'detail'],
})
rel_major_university = rel_major_university.drop_duplicates(subset=['fk_major_id'], keep='first')
rel_major_university.to_json('rel_major_university.json')
# rel_major_university
```
